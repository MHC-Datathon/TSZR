```python
import requests
import pandas as pd
import geopy
from geopy.geocoders import Nominatim
```

```python
# MTA ACE Bus Stop Violations API endpoint 
api = "https://data.ny.gov/resource/kh8p-hcbm.json?$query=SELECT%20violation_id%2C%20vehicle_id%2C%20first_occurrence%2C%20last_occurrence%2C%20violation_status%2C%20violation_type%2C%20bus_route_id%2C%20violation_latitude%2C%20violation_longitude%2C%20stop_id%2C%20stop_name%2C%20bus_stop_latitude%2C%20bus_stop_longitude%2C%20violation_georeference%2C%20bus_stop_georeference%20LIMIT%201500000"

try:
    response = requests.get(api) 
    response.raise_for_status()  
    data = response.json()
    df = pd.DataFrame(data)
except Exception as e:
    print("Error:", e)
```

```python
# splitting first_occurence and last_occurence to date and time
df['first_occurence_date'] = pd.to_datetime(df['first_occurrence']).dt.date
df['last_occurence_date'] = pd.to_datetime(df['last_occurrence']).dt.date

# dropping original columns
df.head()
```

```python
# Save dataframe to CSV file
def df_to_csv(df, filename):
    df.to_csv(filename, index=False)
    print(f"Data saved to {filename}")
    print(f"Total records: {len(df)}")
    print(f"File size: {df.memory_usage(deep=True).sum() / 1024:.2f} KB")
```

```python
csv_filename = 'ace_violations/mta_ace_bus_violations_2025-01_to_2025-08.csv'
df_to_csv(df, csv_filename)
```

```python
congestion_df = pd.read_csv("MTA_Congestion_Relief_Zone_Vehicle_Entries__Beginning_2025_20250923.csv")
congestion_df.head()
```

```python
detection_groups = congestion_df['Detection Group'].unique().tolist()
detection_groups
```

```python
def get_lat_long(location_name):
    """
    Retrieves the latitude and longitude for a given location name.

    Args:
        location_name (str): The name of the location

    Returns:
        tuple: A tuple containing (latitude, longitude) if found, otherwise None.
    """
    # Initialize Nominatim geocoder with a user agent (can be any string)
    geolocator = Nominatim(user_agent="mtacongestion")

    try:
        # Geocode the location name
        location = geolocator.geocode(location_name)

        if location:
            return location.latitude, location.longitude
        else:
            return None
    except Exception as e:
        print(f"Error during geocoding: {e}")
        return None

detection_groups_lat_long = {}
replace_groups = {'West Side Highway at 60th St': 'One Waterline Square, NY', 'FDR Drive at 60th St': '514 E 60th St, NY'}
detection_groups = [group if group not in replace_groups else replace_groups[group] for group in detection_groups]

for group in detection_groups:
    location_name = group
    coordinates = get_lat_long(location_name)

    if coordinates:
        detection_groups_lat_long[group] = coordinates
    else:
        detection_groups_lat_long[group] = None

detection_groups_lat_long
```

```python
for row in congestion_df.itertuples():
    group = row._11  

    group = group if group not in replace_groups else replace_groups[group]
    if group in detection_groups_lat_long and detection_groups_lat_long[group]:
        lat, long = detection_groups_lat_long[group]
        congestion_df.at[row.Index, 'Latitude'] = lat
        congestion_df.at[row.Index, 'Longitude'] = long
    else:
        congestion_df.at[row.Index, 'Latitude'] = None
        congestion_df.at[row.Index, 'Longitude'] = None
```

```python
df_to_csv(congestion_df, 'mta_congestion/mta_congestion_relief_zone_2025-01_to_2025-08.csv')
```
